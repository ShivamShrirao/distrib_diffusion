```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X, y = make_moons(n_samples=1000, noise=0.03, random_state=42)

c1 = X[y==0]
c2 = X[y==1]
plt.scatter(c1[:, 0], c1[:, 1])
plt.scatter(c2[:, 0], c2[:, 1])
plt.show()
```
